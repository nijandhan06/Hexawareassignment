```python
import os
import re
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

# === NLTK Downloads ===
nltk.download('punkt', force=True)
nltk.download('punkt_tab')
nltk.download('stopwords', force=True)
nltk.download('wordnet', force=True)

# === Configuration ===
root_folder = '/content/drive/MyDrive/Usecases'
column_name = 'Description'
software_keywords = set()

# === Initialize NLP tools ===
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

# === Process all files ===
for foldername, _, filenames in os.walk(root_folder):
    for filename in filenames:
        if not (filename.endswith('.csv') or filename.endswith('.xlsx')):
            continue  # Skip non-spreadsheet files

        file_path = os.path.join(foldername, filename)
        print(f"\nğŸ“‚ Processing: {file_path}")

        try:
            # Read CSV
            if filename.endswith('.csv'):
                try:
                    df = pd.read_csv(file_path, encoding='utf-8', on_bad_lines='skip')
                except:
                    df = pd.read_csv(file_path, encoding='ISO-8859-1', on_bad_lines='skip')

            # Read XLSX
            elif filename.endswith('.xlsx'):
                df = pd.read_excel(file_path)

            # Check column
            if column_name not in df.columns:
                print(f" Skipped (missing column): {filename}")
                continue

            # Extract keywords
            for line in df[column_name].dropna():
                tokens = word_tokenize(str(line).lower())
                for word in tokens:
                    clean_word = lemmatizer.lemmatize(word.strip())
                    if (clean_word not in stop_words and
                        len(clean_word) > 2 and
                        clean_word.isalpha()):
                        software_keywords.add(clean_word)

        except Exception as e:
            print(f"Error processing {filename}: {e}")

# === Create Regex Pattern (Optional) ===
regex_pattern = r'\b(?:' + '|'.join(
    re.escape(word) for word in sorted(software_keywords, key=len, reverse=True)
) + r')\b'

# === Save Keywords to CSV ===
keywords_df = pd.DataFrame(sorted(software_keywords), columns=["Keyword"])
keywords_df.to_csv("/content/software_keywords.csv", index=False)
print(f"\nâœ… Collected {len(software_keywords)} keywords")
print("ğŸ”¤ Sample keywords:", list(software_keywords)[:20])
print("ğŸ’¾ Saved to /content/software_keywords.csv")
```
